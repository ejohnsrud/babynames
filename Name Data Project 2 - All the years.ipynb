```python
target = (r'C:\Users\Erin\BirthNamesData')
filelist=os.listdir(target)
filelist
```

```python
#read in all the files (do it as a loop!), then concatenate
df_list = [pd.read_table(file, header=None, names=["Name","Gender","Count"],sep=",") for file in filelist]
big_df = pd.concat(df_list)
```

```python
big_df.head(5)
```

```python
big_df.tail(5)
```

Notice that the lowest name count is five. I'm not 100% on this, but I believe this is because the government is very cautious about releasing any data that could be used to identify a specific person, and names with four or less occurrences is too unique. 

```python
big_df = big_df.groupby(['Name']).sum()
```

```python
big_df
```

```python
big_df = big_df.sort_values(['Count'], ascending=False)
```

```python
big_df
```

```python
big_df=big_df.reset_index()
big_df
```

```python
top_ten = big_df.head(10)
top_ten
```

```python
top_ten.plot.bar(x='Name', y='Count')
```
