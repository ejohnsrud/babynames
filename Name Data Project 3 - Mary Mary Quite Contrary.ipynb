```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

Let's look at the name 'Mary'. 

```python
#First, let's load everything...
target = (r'C:\Users\Erin\BirthNamesData')
filelist=os.listdir(target)
os.chdir(target)
df_list = pd.DataFrame()
for file in filelist:
    foo = pd.read_table(file, header=None, names=['Name','Gender','Count', 'Year'], sep=",")
    foo['Year']=file
    df_list = df_list.append(foo)
```

```python
#Right now the 'Year' column has some unnecessary stringage to it; let's change that. 
df_list['Year'] = df_list['Year'].map(lambda x: x.lstrip('yob').rstrip('.txt'))
```

```python
df_mary = df_list[df_list.Name == 'Mary']
df_mary
```

Anyways, back to all the male Marys. There's a couple of explanations for the men named Mary. First off, the males named Mary could all actually be females whose Social Security applications all had the Gender field mismarked. Secondly, I know that in Catholic tradition Mary or one of its variants is a fairly common middle name; it's not implausible to think that it ended up as first names for some men. Mary could also be short for Meredith, Merriweather, etc. 

Another possibility would depend on whether or not the data is collected from the original Social Security applications or from amended applications. I believe that it is easier to legally change one's first name than it is to legally change one's gender. It's certainly possible that some transgender folks changed their name but didn't legally change their gender, or hadn't as of the time of data collection. I don't know how possible this would be in 1880, but now...

Or it could just be that some parents really wanted a girl and when the baby turned out to be male got a little passive-aggressive with their disappointment. Or, you know, people just give their kids weird/inappropriate names sometimes...
